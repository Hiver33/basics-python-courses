```python
# import du module stat de scipy
from scipy import stats
```

```python
# nous voulons déterminer les valeurs extrêmes et s'il y'en a, les valeurs abérentes
# regles : x < Q1 - 1.5*IQR || x > Q3 + 1.5*IQR

# à parir de cette règles, nous allons construire une fonction de détermination

def outliers_val(data, var_name):
    # calcul de l'IQR, Q1 et Q3
    IQR = stats.iqr(data[var_name])
    Q1 = data[var_name].quantile(0.25)
    Q3 = data[var_name].quantile(0.75)
    
    # calcul de la règle
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    
    # détermination de la dataframe des valeurs etrêmes
    outliers_df = data[(data[var_name] < lower) | (data[var_name] > upper)]
    
    return outliers_df
```

```python
# application de la fonction : 
outliers_val(data = tips_df, var_name = 'tip')
```

### 1 - Fonction de répartion empirique

#### la fonction de répartition empirirque d'une varibale réelle est la fonction X qui a tout X assicie la probabilité d'avoir une valeur inférieur ou égale X
#### $$F(x) = P(X <= x)$$

```python
# nous utiliserons le jeu de données iris qui comprends 50 obs
# ce sont des mesures de longueur faites sur 3 espèces catégorisées 
```

```python
# importation du jeu de données 
iris = sns.load_dataset('iris')
iris.head(5)
```

```python
# afficher la structure de la dataframe
iris.info()
```

```python
# voir les espèces / obs du jeu de données
iris['species'].value_counts()
```

```python
# tracer la fonction de répartition 
x = np.sort(iris[iris.species == 'versicolor']['petal_length'])
x
```

```python
# calcul de la probabilité 1/n
# pour chacune des valeurs, calculons la probabilité 1/n

y = np.arange(1, len(x) + 1) / len(x)
y
```

```python
# cette fonction de répartion attribue la probabilité 1/n a chacun des n observations dans un échantillon
```

```python
# tracé du graphique 
plt.style.use('ggplot')
plt.figure(figsize =(4, 3.5))
plt.plot(x, y)
plt.xlabel('Longueur des pétales')
plt.ylabel('FRE')
plt.title('Iris versicolor')
plt.show()
```

```python
# definotion la fonction de calcul FRE
def fre(data):
    x = np.sort(data)
    y = np.arange(1, len(x) + 1)/len(x)
    return x, y
```

```python
# longueurs des pétales de chaque espèces
# maintenant que nous avons définie une fonction, calcuclons-en les valeurs pour toutes les espèces
x_vers, y_vers = fre(iris[iris.species == 'versicolor']['petal_length'])
x_set, y_set = fre(iris[iris.species=='setosa']['petal_length'])
x_virg, y_virg = fre(iris[iris.species == 'virginica']['petal_length'])
```

### 3 - Utilisation des simulations pour l'estimation de probabilités

#### la simulation est une technique qui permet de modéliser les systèmes plus ou moins complexes. Elle est utilitée pdans plusieurs domaines. La simulation nous donne une solution approximative 

```python
## toujours rapeller les librairies / outilis de manipulations
import pandas as pd
import numpy as np
```

```python
## pour réaliser une simulation : 
# 1 - définir l'ensemble de tous les résultats possibles associés à une variable aléatoires
# 2 - déterminer comment simuler un résultat aléatoire
# 3 - déterminer la règle de succès
# 4 - échantillonner à plusieurs reprises et compter les résultats
# 5 - calculer la fréquence du succès comme une estimation de probabilité
```

```python
## Exemple 1 : 
# soit une expérience consistant à lancer 5 fois une pièce de monnaie 
# il y a 50% de chance d'obtenir 'pile' et 50% de chance d'obtenir 'face'
# un lancé de la pièce est une épreuve de Bernoulli - expérience aléatoire à 2 issues : succès et échec
# -----
# nous allons définir les événements obtenir 'pile' : succès / obtenir 'face' : échec
# l'objectif : obtenir 'N' fois l'expérence 'E',  calculer la probabilité d'obtenir 5 fois 'pile' sur 5 lancés
```

```python
# pour réaliser cette expérience, nous allons générer des nombres réels pseudo-aléatoires compris [0,1]
# si le nombre < 0.5 : 'pile' / si le nombre > 0.5 : 'face'
```

```python
# géérons les nombre pseudo-réels
np.random.seed(1111)
outcomes =  np.random.random(size = 5)
print(outcomes)
```

```python
# compte le nombre de fois qu'on a obtenu 'pile' / <0.5
print(np.sum(outcomes<0.5))
```

```python
# simulation de 'N' expéreince 'E'
N = 1000000 # le nb de repétition de l'expérience

n_all_piles = 0 # nb fois 'E' a donnée 'pile'

for _ in range(N):
    results = np.random.random(size = 5)
    piles = results<0.5
    n_piles = np.sum(piles)

    if n_piles == 5:
        n_all_piles += 1

prob_all_piles = n_all_piles / N

print("la probabilité d'obtenir 5 fois 'pile' sur les 5 lancés est de :", prob_all_piles*100, '%')
```

```python
## Exemple 2 : 
# on lance deux (2) dés numérotés de 1-6. tous les numéros ont la meme probabilité d'apparition (distribution uniforme)
# on gagne si les deux (2) dés affiche le meme numéro et on perd dans le cas contraire

# aprsè avoir réalisé le jeu une (1) fois, on va repetrer 'N' fois afin d'estimer la probabilité de gagner
```

```python
# réalisons unue fois le jeu : affactation multiples
de, probs, n_de = [1,2,3,4,5,6], [1/6,1/6,1/6,1/6,1/6,1/6], 2
outcomes = np.random.choice(de, size = n_de, p = probs)
outcomes
```

```python
# forcons la condition pour gagner 
if outcomes[0] == outcomes[1]:
    issue = 'gagné'
else:
    issue = 'perdu'
print('les 2 dés affichent :', outcomes[0], 'et', outcomes[1],'\ndonc on a', issue)
```

```python
# jouons plusieurs fois le jeu pour calculer la fréquance de probabilité
N = 1000000 
n_gagne = 0
for i in range(N):
    outcomes = np.random.choice(de, size = n_de, p = probs)
    if outcomes[0] == outcomes[1]:
        n_gagne +=1
prob_gagne = n_gagne/N

print('la probabilité de gagner à ce jeu de dé est de :', prob_gagne*100, '%')
```

```python
## Exercice 3 :
# on dispose d'une urne contenant 7 boules blanches et 6 boules nooirs
# on tire successivement et sans remise 4 boules
# on considère comme succès : l'événement la 1ere et la 3e boules soient blanches et la 2e et 4e boules soient noires
# réalisons la simulation
```

```python
## corrigé :
# déclaration des variables
n_succes = 0
N = 1000
urne = ['b','b','b','b','b','b','b','n','n','n','n']
```

```python
# réalisons l'expérience
for i in range(N):
    tirage = np.random.choice(urne, replace = False, size = 4)
    if ((tirage[0] == 'b' and tirage[2] == 'b') and (tirage[1] == 'n' and tirage[3] == 'n')):
        n_succes += 1
prob_succes = n_succes/N
print('la probabilité de succès est estimé à :', prob_succes*100, '%')
```

# Chapitre 5 : Bootstrapping : une méthode de rééchantillonnage

```python
# technique de reévalutaion de l'échantillon
# ici on pense / parle de l'intervalle de confiance
# la méthode consiste à refaire l'échantillonnage mais avec remise 
# on repete plusieurs fois la simulation afin d'obtenir un grand nombre stats récapitulatives 
```

### 1 - Echantillon bootstrap

```python
# nous allons utiliser les données de précipation de Dallas
df = pd.read_csv('Dallas weather data.csv', index_col = 'Year')
df = pd.DataFrame(df)
```

```python
# ordoner les données en croissance suivant les années
df = df.sort_values('Year')
df.head()
```

```python
# vérifier l'existance des valeurs NaN
df.isna().sum()
```

```python
# remplaçons / imputons les NaN par la moyenne de la colonne
# convertissons les colonnes en valeurs numériques
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors = 'coerce')
df.fillna(df.mean(), inplace = True)
df.head()
print(type(df))
```

```python
# selectionnons 2 colonnes Year et Total
df = df.reset_index()[['Year', 'Total']]
df.head()
```

```python
# strucure de données 
print(df.info())

print('-- statistique descriptive --\n',df.describe())
```

```python
# les stats révèle la moyenne à 33.34 mm. mais, à quel niveau de certitude ?
# on pense qu'à cet étape, il est prématuré de confirmer cette statistiques, 
# car l'on peut se permettre d'imputer les fluctuations des valeurs au hasard du à l'échantillonnage
# il est donc interressant pour nous de fixer la plage / l'intervalle de confiance de cette statistique
# c'est ainsi tout le sens de la méthode bootstrapping - elle nous permettra de générer 'n' échantillons pour autant de stats
# ainsi, l'on parviendra à situer avec confiance les plages de moyennes pour cette étude
```

```python
# réévaluation de l'échantillon : créer un ou plusieurs échantillons avec le jeu de données 
# pour se rapprocher des résultats réels de l'expérience, il faut que l'échantillon soit de meme taille que la population réelle
```

```python
# créer un échantillon
boot_sample = np.random.choice(df['Total'], size = len(df))
print(boot_sample)

if len(boot_sample) == len(df['Total']):
    print("\nrelevé d'origine et échantillon bootstrap sont de même taille :", len(boot_sample))
else:
    print("revoir la méthode de rééchantillonage")
```

```python
# calcul de la moyenne de l'échantillon
np.mean(boot_sample)
```

```python
# calcul de l'écart-type
np.std(boot_sample)
```

```python
# Rappel de fonction de répartition empirique
def fre(data):
    x = np.sort(data)
    y = np.arange(1, len(x)+1)/len(x)
    return x, y
```

```python
# générons 1000 échantillons bootstrap
# créyons une liste qui va enregistrer à chaque itération la moyenne de l'échantillon bootstrapping
plt.figure(figsize =(5, 3.5))
mean_boots = []

for _ in range(1000):
    bs_sample = np.random.choice(df['Total'], size = len(df))
    mean_boots.append(np.mean(bs_sample))
    x_boot, y_boot = fre(bs_sample)
    _ = plt.plot(x_boot, y_boot, marker = '.', linestyle = 'none', color = 'blue', alpha = 0.1)

# fonction de repartion (fre)
x,y = fre(df['Total'])
_ = plt.plot(x,y, marker = '.', color = 'red')
_ = plt.xlabel('Précipitation annuelle')
_ = plt.ylabel('FRE')
```

```python
# les moyennes des échantillon en blue désigne l'intervalle / plage de moyenne
# les données réelles sont revellées par la ligne rouge
# ainsi, on comprends que la moyennes peut varier suivant l'intervalles tracés par la méthode de rééchantillonage
```

### 2 - Intervalles de confiance bootstrap

```python
# maintenant que nous avons avons plus ou moins l'idée sur la plage de moyennes des des précipitation,
# nous pouvons calculer l'intervalle de confiance

# l'intervalle de confiance est l'intervalle où se situerait les données réccapitulatives 
# quand l'échantillonnage est repeté n fois
```

```python
# calculons l'intervalle de confiance à 95% de notre stat bootstrap
# il faudra calculer les percentiles de la liste mean_boots (moyenne des échantillons)
np.percentile(mean_boots, [2.5, 97.5])
```

```python
# ceci veut dire :  pour n'importe quel échantillon des données précipitation à Dallas,
# nous sommes sûr à 95% que la moyennes se situerait entre [31.72842549, 34.89902148]
```

```python
# vérifions notre intervalle avec les données originelles 'df'
np.mean(df['Total'])
```

```python
# fonction de génération de plusieurs réplique bootstrap
def compute_boots_repliqs(data, function, size =1):
    boots_repliq = np.empty(size)

    for i in range(size):
        boots_repliq[i] = function(np.random.choice(data, len(data)))
    return boots_repliq
```

```python
# utilisons cette fonction pour calculer l'intervalle de confiance à 95%
# et voir si nous avons le meme résultat que précédemment

np.percentile(compute_boots_repliqs(df['Total'], np.mean, size = 1000), [2.5, 97.5])
```

```python
# l'on peut se permettre aussi de calculer l'intervalle à 95% des autres stat récapitulatives 
```

```python
# créyons le graphique de la repliques des échantillon bootsstrap
plt.figure(figsize =(6, 4))
sns.displot(compute_boots_repliqs(df['Total'], np.mean, size = 1000),
           kde = True, edgecolor = 'none',
           height = 3.5, aspect = 1.5)
plt.xlabel('Précipitation annuelle en inches')
plt.title('Distribution des répliques bootstrap de la précipitation des moyennes',
         fontsize = 11)
```

```python
# ce graphique montre clairement que la distribution des repliques bootstrapping 
# suit quasiment une loi normale empirique
```
